## Analisando Vendas por Tempo

```python
vendas['data_venda'] = pd.to_datetime(vendas['data_venda'])
vendas.head()
```

```python
soma_valor_data = vendas.groupby('data_venda').sum()['valor_venda'].to_frame().reset_index().rename(columns = {'valor_venda': 'soma_valor_venda'})
media_valor_data = vendas.groupby('data_venda').mean()['valor_venda'].to_frame().reset_index().rename(columns = {'valor_venda': 'media_valor_venda'})
mediana_valor_data = vendas.groupby('data_venda').median()['valor_venda'].to_frame().reset_index().rename(columns = {'valor_venda': 'mediana_valor_venda'})
```

```python
soma_vendas_ano_2012 = vendas.query("data_venda >= '2012-01-01' and data_venda <='2012-12-31'")
soma_vendas_ano_2013 = vendas.query("data_venda >= '2013-01-01' and data_venda <='2013-12-31'")
soma_vendas_ano_2014 = vendas.query("data_venda >= '2014-01-01' and data_venda <='2014-12-31'")
soma_vendas_ano_2015 = vendas.query("data_venda >= '2015-01-01' and data_venda <='2015-12-31'")
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
grafico = sns.lineplot(x = 'data_venda', y = 'valor_venda', data = soma_vendas_ano_2012, palette = 'Greens_d')
grafico.figure.set_size_inches(16,8)
grafico.set_title('Vendas no Ano de 2012', fontdict={'fontsize':18, 'fontweight':'bold'})
grafico.set_xlabel('Tempo', fontsize = 12)
grafico.set_ylabel('Valor Venda', fontsize = 12)
grafico = grafico
```
